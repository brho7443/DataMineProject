# Respiratory Analysis

The purpose of this document is to analyze the various changes in respiratory conditions according to medicaid before, during, and after the COVID-19 pandemic. The reason why we need this data is to see how respiratory illnesses have changed over time, and if there's a correlation between air quality and the various diseases.

## Data Processing/ Cleaning

As always, lets take care of our data set to make sure that it contains no misleading information/ other bad entries.

```python
#load data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


healthData = pd.read_csv('/home/jovyan/DataMiningProject/DataMineProject/Data_Mine_Project/CSV Files/Respiratory Conditions Per Medicaid.csv')
```

```python
#check for NULL values
death_nulls = healthData.isnull().sum().sum()  

# print total count
print(f"Total null values in data set: {death_nulls}")

# Count the number of null values in each column
null_counts = healthData.isnull().sum()

# Print the result
print(null_counts)
```

```python
# Drop the final column (by index)
healthData = healthData.drop(healthData.columns[-1], axis=1)

# Check the new dataset
print(healthData.head())

#check for NULL values
death_nulls = healthData.isnull().sum().sum()  

# print total count
print(f"Total null values in data set: {death_nulls}")

# Count the number of null values in each column
null_counts = healthData.isnull().sum()

# Print the result
print(null_counts)
```

## Data Analysis

With the dataset cleaned, let's find out how many unqiue diseases there are in the set, and get their metrics for future evaluations.

```python
# Get unique values from a specific column (replace 'column_name' with the actual column name)
conditions = healthData['Condition'].unique()

# Print the unique values
print(conditions)
```

```python
def getBeneficiaryCounts(diseaseName, year):
    tempData = healthData[
    (healthData['Year'] == year) & 
    (healthData['Condition'] == diseaseName)]['BeneficiaryCount']
    
    vals = []
    for i in tempData:
        # Remove commas and check if the value is numeric
        cleaned_value = i.replace(',', '')
        if cleaned_value.isdigit():  # Check if the value is numeric
            vals.append(int(cleaned_value))  # Convert to integer before appending
            
    total = sum(vals)
    return(total)
```

```python
diseaseNumbersNamesAndYears = []
diseaseYears = []
diseaseNames = []
diseaseCounts = []

for i in range(2018, 2023):
    print(f"{i} Data")
    for disease in conditions:
        count = getBeneficiaryCounts(disease, i)
        print(f"{disease}:  {count}")
        diseaseNumbersNamesAndYears.append((disease, i, count))
        diseaseYears.append(i)
        diseaseNames.append(disease)
        diseaseCounts.append(count)
    print("\n\n")
```

```python
bronchitisValues = []
distressSyndrome = []
bronchitisNos = []
influenza = []
respInfec = []
pneumonias = []
respNos = []

for i in range(2018, 2023):
    count = getBeneficiaryCounts("Acute bronchitis", i)
    bronchitisValues.append(count)
    
    count = getBeneficiaryCounts("Acute respiratory distress syndrome", i)
    distressSyndrome.append(count)
    
    count = getBeneficiaryCounts("Bronchitis NOS", i)
    bronchitisNos.append(count)
    
    count = getBeneficiaryCounts("Influenza", i)
    influenza.append(count)
    
    count = getBeneficiaryCounts("Lower or acute respiratory infection", i)
    respInfec.append(count)
    
    count = getBeneficiaryCounts("Pneumonia", i)
    pneumonias.append(count)
    
    count = getBeneficiaryCounts("Respiratory infection NOS", i)
    respNos.append(count)

print(bronchitisValues)
print(distressSyndrome)
print(bronchitisNos)
print(influenza)
print(respInfec)
print(pneumonias)
print(respNos)
```

```python
plt.title('Total Disease Counts (United States)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Disease Counts (Tens of Millions)', fontsize=12)


uniqueYears = np.arange(2018, 2023)


plt.plot(uniqueYears, bronchitisValues, label='Bronchitis', color='r', marker='o')  # Line plot with markers
plt.plot(uniqueYears, distressSyndrome, label='Acute respiratory distress syndrome', color='g', marker='o')  # Line plot with markers
plt.plot(uniqueYears, bronchitisNos, label='Bronchitis NOS', color='b', marker='o')  # Line plot with markers
plt.plot(uniqueYears, influenza, label='Influenza', color='c', marker='o')  # Line plot with markers
plt.plot(uniqueYears, respInfec, label='Lower or acute respiratory infection', color='m', marker='o')  # Line plot with markers
plt.plot(uniqueYears, pneumonias, label='Pneumonia', color='y', marker='o')  # Line plot with markers
plt.plot(uniqueYears, respNos, label='Respiratory infection NOS', color='brown', marker='o')  # Line plot with markers


# Adding a grid for better readability
plt.grid(True)
    
# Show legend
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=3)

# Run only when you need a new copy of the image. 
#plt.savefig('/home/jovyan/DataMiningProject/DataMineProject/Data_Mine_Project/Visuals/Total Disease Counts.png')

plt.show()
```
