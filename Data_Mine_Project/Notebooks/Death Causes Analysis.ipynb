# Introduction to this notebook

It's probably useful to first understand how the United States compares in terms of mortality rates and overall health over the years. That said, I would love to know how we as Americans are doing in terms of mortality and death.  

```python
#load data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


deathData = pd.read_csv('/home/jovyan/DataMiningProject/DataMineProject/Data_Mine_Project/CSV Files/LeadingCausesOfDeathInUSA.csv')
```

```python
#check for NULL values
death_nulls = deathData.isnull().sum().sum()  

# print total count
print(f"Total null values in death Data: {death_nulls}")
```

```python
# Get years
uniqueYears = np.arange(1999,2018)

allDeaths = deathData[
    (deathData['113 Cause Name'] == 'All Causes') &
    (deathData['Cause Name'] == 'All causes') &
    (deathData['State'] == 'United States')] ['Deaths']
allDeaths = allDeaths.iloc[::-1]

print(allDeaths)
print(len(allDeaths))
print(len(uniqueYears))
```

```python
plt.title('Total Deaths from All Causes (United States)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Deaths', fontsize=12)

plt.plot(uniqueYears, allDeaths, label='Total Deaths from All Causes (millions) ', color='b', marker='o')  # Line plot with markers

    
# Adding a grid for better readability
plt.grid(True)
    
# Show legend
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=3)

# Run only when you need a new copy of the image. 
#plt.savefig('/home/jovyan/DataMiningProject/DataMineProject/Data_Mine_Project/Visuals/AllDeathsAllCauses.png')


plt.show()
```

## Analysis

Given the steep increase of deaths from 2010 onwards, it becomes pretty clear that there's a reason why so many people are dying so suddenly in the United States. Having said that, we want to find out why this is the case, and -- more importantly -- if this has something to do with air quality throughout the country. 

```python
infantDeaths = pd.read_csv('/home/jovyan/DataMiningProject/DataMineProject/Data_Mine_Project/CSV Files/Childhood Mortality.csv')
#check for NULL values
death_nulls = deathData.isnull().sum().sum()  

# print total count
print(f"Total null values in death Data: {death_nulls}")
```

```python
# Get years
uniqueYears = np.arange(1900,2019)

# 1-4 Years
allInfantDeaths = infantDeaths[
    (infantDeaths['Age Group'] == '1-4 Years')]['Death Rate']

# 5-9 Years
allToddlerDeaths = infantDeaths[
    (infantDeaths['Age Group'] == '5-9 Years')]['Death Rate']


# 10-14 Years
allPreteenDeaths = infantDeaths[
    (infantDeaths['Age Group'] == '10-14 Years')]['Death Rate']

# 15-19 Years
allTeenageDeaths = infantDeaths[
    (infantDeaths['Age Group'] == '15-19 Years')]['Death Rate']



print(len(allInfantDeaths))
print(len(uniqueYears))
print(len(allToddlerDeaths))
print(len(allPreteenDeaths))
print(len(allTeenageDeaths))
```

```python
plt.title('Mortality Rates', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Mortality Rate', fontsize=12)

plt.plot(uniqueYears, allInfantDeaths, label='Infant Mortality Rate ', color='b', marker='.')  # Infants
plt.plot(uniqueYears, allToddlerDeaths, label='Toddler Mortality Rate ', color='r', marker='.') # Toddlers
plt.plot(uniqueYears, allPreteenDeaths, label='Preteen Mortality Rate ', color='g', marker='.') # Preteens
plt.plot(uniqueYears, allTeenageDeaths, label='Teenager Mortality Rate ', color='m', marker='.') # Teenagers

    
# Adding a grid for better readability
plt.grid(True)
    
# Show legend
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=2)

# Run only when you need a new copy of the image. 
#plt.savefig('/home/jovyan/DataMiningProject/DataMineProject/Data_Mine_Project/Visuals/AllMortalityRates.png')

plt.show()
```
