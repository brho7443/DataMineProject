```python
import matplotlib.pyplot as plt


# Get all fuel values 
all_fuels_data = emissionsData[emissionsData['fuel-name'] == 'All Fuels']
all_produced_emissions = all_fuels_data['value']
all_years = all_fuels_data['year']


# Get years
uniqueYears = np.arange(1970,2022)


# Get total co2 from all sectors and their years
filtered_values = emissionsData[
    (emissionsData['state-name'] == 'United States') &
    (emissionsData['sector-name'] == 'Total carbon dioxide emissions from all sectors') &
    (emissionsData['fuel-name'] == 'All Fuels')] ['value']


plt.title('Total CO2 Emissions from All Sectors (United States)', fontsize=14)
plt.xlabel('uniqueYears', fontsize=12)
plt.ylabel('Emissions (CO2 in Metric Tons)', fontsize=12)

plt.plot(uniqueYears, filtered_values, label='Total CO2 Emissions', color='b', marker='o')  # Line plot with markers

    
# Adding a grid for better readability
plt.grid(True)
    
# Show legend
plt.legend()


# Run only when you need a new picure
# plt.savefig('/home/jovyan/DataMiningProject/DataMineProject/Data_Mine_Project/Visuals/TotalCO2FromAllSectors.png')


plt.show()
```

```python
coalValues = emissionsData[
    (emissionsData['state-name'] == 'United States') &
    (emissionsData['sector-name'] == 'Total carbon dioxide emissions from all sectors') &
    (emissionsData['fuel-name'] == 'Coal')] ['value']

petrolValues = filtered_values = emissionsData[
    (emissionsData['state-name'] == 'United States') &
    (emissionsData['sector-name'] == 'Total carbon dioxide emissions from all sectors') &
    (emissionsData['fuel-name'] == 'Petroleum')] ['value']

gasValues = filtered_values = emissionsData[
    (emissionsData['state-name'] == 'United States') &
    (emissionsData['sector-name'] == 'Total carbon dioxide emissions from all sectors') &
    (emissionsData['fuel-name'] == 'Natural Gas')] ['value']

plt.title('Total CO2 Emissions from various sources US)', fontsize=14)
plt.xlabel('Years', fontsize=12)
plt.ylabel('Emissions (CO2 in Metric Tons)', fontsize=12)

plt.plot(uniqueYears, coalValues, label='Coal Emissions', color='r', marker='o')
plt.plot(uniqueYears, petrolValues, label='Petroleum Emissions', color='g', marker='o')
plt.plot(uniqueYears, gasValues, label='Gas Emissions', color='y', marker='o')

    
# Adding a grid for better readability
plt.grid(True)
    
# Show legend
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=3)

# Run only when you need a new picure
#plt.savefig('/home/jovyan/DataMiningProject/DataMineProject/Data_Mine_Project/Visuals/TotalCO2VariousSources.png')

plt.show()
```
